### Resumo do documento "Concepts of Cybersecurity"

#### Definição de Segurança Computacional
- **Segurança Computacional**: Medidas e controles que garantem a confidencialidade, integridade e disponibilidade dos ativos do sistema de informação, incluindo hardware, software e dados processados, armazenados e comunicados.

#### Perguntas Fundamentais de Segurança
1. Quais ativos precisam ser protegidos?
2. Como esses ativos são ameaçados?
3. O que podemos fazer para combater essas ameaças?

#### Objetivos Fundamentais de Segurança (Triângulo CIA)
- **Confidencialidade**: Restringir o acesso e a divulgação de informações a apenas aqueles autorizados.
- **Integridade**: Garantir que a informação não seja modificada ou destruída sem autorização.
- **Disponibilidade**: Assegurar o acesso e uso de informações de maneira oportuna e confiável.

#### Outros Conceitos de Segurança
- **Autenticidade**: Verificação da identidade e origem das informações.
- **Responsabilidade**: Ações de uma entidade devem ser rastreáveis de forma única a ela.

#### Segurança, Segurança e Confiabilidade
- **Segurança**: Envolve ações intencionais que tentam explorar vulnerabilidades.
- **Segurança**: Lida com ações não intencionais, como problemas de confiabilidade de rede.
- **Confiabilidade**: Propriedade de um sistema que permite que se confie nos serviços fornecidos.

#### Comparação: Segurança Física vs. Digital
- **Ativos**: Ouro (físico) vs. Informação (digital).
- **Metas**: Evitar roubo físico vs. garantir confidencialidade, integridade e disponibilidade.
- **Proteção**: Cadeados e guardas armados vs. criptografia, firewalls e auditorias de segurança.

#### Níveis de Ataque
- **Humano**: Exemplo, se passar por suporte de TI.
- **Aplicação**: Exemplo, gerenciamento de senhas vulnerável.
- **Sistema Operacional**: Exemplo, overflow de buffer.
- **Hardware**: Exemplo, ler disco rígido abrindo o case do computador.

#### Impacto de uma Violação de Segurança
- **Baixo**: Efeito limitado nas operações organizacionais.
- **Moderado**: Efeito adverso sério, mas não catastrófico.
- **Alto**: Efeito severo ou catastrófico, incluindo perda de vida ou lesões graves.

#### Terminologia de Segurança Computacional
- **Adversário (Agente de Ameaça)**: Entidade que realiza atividades prejudiciais.
- **Ataque**: Atividade maliciosa que visa coletar, interromper, degradar ou destruir recursos do sistema de informação.
- **Contramedida**: Dispositivo ou técnica que visa impedir atividades indesejáveis.
- **Risco**: Medida da extensão em que uma entidade é ameaçada por uma circunstância ou evento potencial.
- **Política de Segurança**: Conjunto de critérios para fornecer serviços de segurança.

#### Desafios da Segurança Computacional
1. Complexidade inerente.
2. Consideração de ataques potenciais em todos os mecanismos de segurança.
3. Procedimentos frequentemente contra-intuitivos.
4. Decisão sobre onde usar mecanismos de segurança.
5. Envolvimento de múltiplos mecanismos ou protocolos.
6. Batalha contínua entre perpetradores e defensores.
7. Tendência de subestimar os benefícios de investimentos em segurança.
8. Necessidade de monitoramento constante.
9. Segurança como uma reflexão tardia no design do sistema.
10. Percepção de que a segurança forte impede a operação eficiente.

- **Design aberto**
- **Separação de privilégios**

- **Mecanismo comum mínimo**
- **Aceitabilidade psicológica**
- **Modularidade**
- **Camadas**
- **Menor surpresa**

Este resumo fornece uma visão geral dos conceitos fundamentais de segurança cibernética, incluindo objetivos principais, desafios e princípios de design necessários para proteger sistemas de informação de ameaças e ataques.
