### Pontos-chave do documento "Managing Security"

#### Decisão Gerencial na Segurança
- A implementação de medidas de segurança é uma decisão gerencial.
- Medidas técnicas de segurança devem trabalhar junto com medidas organizacionais.
- Decisões de gestão devem ser baseadas em análises de riscos e ameaças atuais.

#### Motivação para Ataques
- Ganho financeiro.
- Desafio técnico (e reconhecimento).
- Vingança.
- Vandalismo.
- Hacktivismo.
- Terrorismo.
- Guerra cibernética.

#### Perfil e Competências dos Atacantes
- **Técnico**: Sistemas operacionais, redes, software, criptografia, hardware.
- **Não-técnico**: Engenharia social.
- **Outros campos**: Abertura de fechaduras.
- **Exemplos de atacantes**: Solitários, "script kiddies", criminosos profissionais, governos, provedores de serviços de internet, insiders.

#### Segurança como um Problema de Pessoas
- A segurança não pode ser resolvida apenas pela tecnologia.
- Deve envolver o sistema legal, gestão organizacional e especialistas técnicos.
- Todos os usuários devem cumprir as políticas e práticas de segurança da organização.

#### Segurança e a Organização
- Segurança deve ser abordada desde o início em cada etapa de desenvolvimento de serviços e produtos de TI.
- Treinamento regular de segurança para desenvolvedores.
- Conscientização dos desenvolvedores sobre o ambiente de implantação e regulamentações específicas.

#### Políticas de Segurança
- Definem objetivos de segurança de uma organização.
- Regulam acesso a instalações, documentos, transações comerciais, formatos de senha e intervalos de renovação.
- **Política de segurança organizacional**: Leis, regras e práticas que regulam a gestão, proteção e distribuição de recursos.
- **Política de segurança automatizada**: Restrições e propriedades que especificam como um sistema computacional evita uso indevido dos recursos.

#### Medição da Segurança
- Medir a segurança antes e depois de implementar novos mecanismos.
- Medições objetivas (e.g., número de portas abertas) e subjetivas (e.g., reputação de uma empresa).
- Medição do número de falhas de segurança e superfície de ataque (número de interfaces externas e instruções perigosas).

#### Medindo a Segurança de um Sistema
- Analisar configurações atuais, número de contas com privilégios de sistema, contas com senhas fracas, número de portas abertas e vulnerabilidades conhecidas.
- Medir a conectividade dos nós em uma rede para avaliar a propagação de ataques.

#### Padrões de Segurança
- Alguns setores possuem normas de gestão de segurança prescritivas (e.g., setor financeiro).
- Outros padrões são códigos de melhores práticas (e.g., ISO 27002).

#### ISO 27002
- **Política de segurança**: Direção e suporte da gestão em questões de segurança.
- **Organização da segurança da informação**: Responsabilidades de segurança dentro da empresa.
- **Gestão de ativos**: Conhecimento dos ativos e seu valor.
- **Segurança de recursos humanos**: Fontes de insegurança dentro do pessoal.
- **Segurança física e ambiental**: Medidas de segurança física (e.g., cercas, portas trancadas).
- **Gestão de comunicações e operações**: Manutenção da segurança na gestão diária de sistemas de TI e processos de negócios.
- **Controle de acesso**: Aplicação de controle de acesso a dados, serviços e computadores.
- **Aquisição, desenvolvimento e manutenção de sistemas de informação**: Consideração de questões de segurança no desenvolvimento de sistemas de TI.
- **Gestão de incidentes de segurança da informação**: Preparação para lidar com incidentes de segurança prontamente.
- **Gestão de continuidade de negócios**: Medidas para lidar com falhas ou desastres.
- **Conformidade**: Cumprimento de obrigações legais, regulamentares e contratuais, além das políticas de segurança organizacional.

Este resumo cobre os principais pontos sobre a gestão da segurança, motivação e perfil dos atacantes, a necessidade de abordar a segurança como um problema humano e organizacional, bem como os padrões de segurança e medidas de conformidade.
